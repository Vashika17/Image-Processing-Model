```python
#Broadcasting (For Different Sized Arrays)
ar=np.array([
      [1,2,3],  # 1 2 3
      [4,5,6],
      [7,8,9]    #(3,3) (1,3) atleast one '1' in column/row , same no.of row /column in both array
    ])           #(3,3) (3,2)-> not applicable
br=np.array([1,2,3])
print(ar*br)
```

```python
ar=np.array([
      [1,2,3], #1
      [4,5,6], #2
      [7,8,9]  #3
    ])
br=np.array([[1],[2],[3]])  #(3,3) (3,1)
print(br)
print(ar*br)
```
